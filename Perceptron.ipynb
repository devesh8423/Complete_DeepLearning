# Draw Back of Perceptron is Not good performance on Non  Linear Data


```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.linear_model import Perceptron
```

```python
p=Perceptron()
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y.values,clf=p,legend=2);
```
