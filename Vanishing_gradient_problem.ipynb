```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import keras
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
```

```python
x,y=make_moons(n_samples=250,noise=0.05,random_state=45)
```

# **The vanishing gradient problem is a training issue in deep neural networks where gradients (the signals used to update weights) become extremely small as they propagate backward through the network, causing early layers to update very slowly or not at all. This happens when many layers are used and especially when activation functions with low derivatives, like sigmoid, are present. The problem results in poor learning, as early layers fail to learn complex features, and the network struggles to converge**

```python
plt.scatter(x[:,0],x[:,1],c=y,s=100)
plt.show();
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model=Sequential()
model.add(Dense(10,activation="sigmoid",input_dim=2))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(10,activation="sigmoid"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.get_weights()[0]
```

```python
old_weights=model.get_weights()[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
new_weights=model.get_weights()[0]
new_weights
```

```python
old_weights
```
